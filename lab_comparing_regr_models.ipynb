```python
# In this final lab, we will model our data. Import sklearn train_test_split and separate the data.
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np


categorical = pd.read_csv('categorical.csv')
numerical = pd.read_csv('numerical_data.csv')

data = pd.concat([categorical, numerical], axis=1)

data.head()
```

```python
data['policy'] = data['policy'].astype(str)
```

```python
# Separate X_train and X_test into numerical and categorical (X_train_cat , X_train_num , X_test_cat , X_test_num)

y = data['total_claim_amount']
X = data.drop(['total_claim_amount', 'customer'], axis=1) 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
X_train_num = X_train.select_dtypes(include = np.number)
X_train_cat = X_train.select_dtypes(include=[np.object])

X_test_num = X_test.select_dtypes(include = np.number)
X_test_cat = X_test.select_dtypes(include=[np.object])
```

```python
# Use X_train_num to fit scalers. Transform BOTH X_train_num and X_test_num.

from sklearn.preprocessing import MinMaxScaler

transformer = MinMaxScaler().fit(X_train_num)  
X_train_norm = transformer.transform(X_train_num)  
X_train_num_norm = pd.DataFrame(X_train_norm, columns=X_train_num.columns)
```

```python
# Great! Now define a function that takes a list of models and train (and tests) 
# them so we can try a lot of them without repeating code.


def test_models(x, y):
    from sklearn.linear_model import Ridge
    from sklearn.metrics import r2_score
    from sklearn.linear_model import LinearRegression, LogisticRegression
    from sklearn.metrics import precision_score, recall_score, f1_score, r2_score
    from sklearn.neighbors import KNeighborsRegressor
    from sklearn.neural_network import MLPRegressor
    
    lm = LinearRegression()
    lm.fit(x, y)
    lm_preds = lm.predict(x)
    print('Linear regression score:', r2_score(y, lm_preds))
    
    ridge = Ridge(alpha=1.0) 
    ridge.fit(x, y)
    ridge_preds = ridge.predict(x)
    print('Ridge regression score:', r2_score(y, ridge_preds))
    
    neighborsRegressor_regressor = KNeighborsRegressor(n_neighbors=5)  
    neighborsRegressor_regressor.fit(x, y)
    neighborsRegressor_preds = neighborsRegressor_regressor.predict(x)
    print('KNeighborsRegressor score:', r2_score(y, neighborsRegressor_preds))
    
    mlp_regressor = MLPRegressor(hidden_layer_sizes=(100, 50), max_iter=500)  
    mlp_regressor.fit(x, y)
    mlp_preds = mlp_regressor.predict(x)
    print('MLPRegressor score:', r2_score(y, mlp_preds))
    

test_models(X_train_one_hot_encoded, y_train)
```
